```python
def create_sol_tables(response_json):
    # Create tables for each sol
    for sol_key, sol_data in response_json.items():
        print(f"\n--- Sol {sol_key} Summary ---")
        
        # Basic Sol Information Table
        basic_info = {
            'First UTC': sol_data['First_UTC'],
            'Last UTC': sol_data['Last_UTC'],
            'Season': sol_data['Season'],
            'Northern Season': sol_data['Northern_season'],
            'Southern Season': sol_data['Southern_season']
        }
        basic_df = pd.DataFrame.from_dict(basic_info, orient='index', columns=['Value'])
        display(basic_df)
        
        # Temperature (AT) Table
        at_data = sol_data['AT']
        at_df = pd.DataFrame.from_dict(at_data, orient='index', columns=['Value'])
        at_df.index.name = 'Temperature Statistic'
        print("\nAir Temperature (AT) Statistics:")
        display(at_df)
        
        # Horizontal Wind Speed (HWS) Table
        hws_data = sol_data['HWS']
        hws_df = pd.DataFrame.from_dict(hws_data, orient='index', columns=['Value'])
        hws_df.index.name = 'Wind Speed Statistic'
        print("\nHorizontal Wind Speed (HWS) Statistics:")
        display(hws_df)
        
        # Pressure (PRE) Table
        pre_data = sol_data['PRE']
        pre_df = pd.DataFrame.from_dict(pre_data, orient='index', columns=['Value'])
        pre_df.index.name = 'Pressure Statistic'
        print("\nPressure (PRE) Statistics:")
        display(pre_df)
        
        # Wind Direction (WD) Most Common Table
        wd_most_common = sol_data['WD']['most_common']
        wd_df = pd.DataFrame.from_dict(wd_most_common, orient='index', columns=['Value'])
        wd_df.index.name = 'Wind Direction Characteristic'
        print("\nMost Common Wind Direction:")
        display(wd_df)
        
        print("\n" + "="*50)
```

```python
# Call the function with your response JSON
create_sol_tables(response.json())
```

```python
def export_sol_tables_to_markdown(response_json, output_file='mars_weather_data.md'):
    """
    Export Mars weather data tables to a Markdown file with error handling
    
    Args:
        response_json (dict/list): JSON response from Mars API
        output_file (str): Path to output Markdown file
    """
    # If response_json is a list, convert to dict if possible
    if isinstance(response_json, list):
        response_json = {str(i): sol for i, sol in enumerate(response_json)}
    
    with open(output_file, 'w') as md_file:
        # Write header
        md_file.write("# Mars Weather Data Analysis\n\n")
        md_file.write(f"*Generated on: {pd.Timestamp.now().strftime('%Y-%m-%d %H:%M:%S')}*\n\n")
        
        # Iterate through each sol
        for sol_key, sol_data in response_json.items():
            md_file.write(f"## Sol {sol_key} Summary\n\n")
            
            # Safely extract data with default values
            def safe_get(data, *keys, default='N/A'):
                for key in keys:
                    try:
                        data = data[key]
                    except (KeyError, TypeError):
                        return default
                return data
            
            # Basic Sol Information
            md_file.write("### Basic Information\n\n")
            md_file.write("| Attribute | Value |\n")
            md_file.write("|-----------|-------|\n")
            
            basic_info = {
                'First UTC': safe_get(sol_data, 'First_UTC'),
                'Last UTC': safe_get(sol_data, 'Last_UTC'),
                'Season': safe_get(sol_data, 'Season'),
                'Northern Season': safe_get(sol_data, 'Northern_season'),
                'Southern Season': safe_get(sol_data, 'Southern_season')
            }
            
            for key, value in basic_info.items():
                md_file.write(f"| {key} | {value} |\n")
            md_file.write("\n")
            
            # Temperature Statistics
            md_file.write("### Air Temperature (AT) Statistics\n\n")
            md_file.write("| Statistic | Value |\n")
            md_file.write("|-----------|-------|\n")
            at_data = safe_get(sol_data, 'AT', default={})
            for key, value in at_data.items():
                md_file.write(f"| {key} | {value} |\n")
            md_file.write("\n")
            
            # Wind Speed Statistics
            md_file.write("### Horizontal Wind Speed (HWS) Statistics\n\n")
            md_file.write("| Statistic | Value |\n")
            md_file.write("|-----------|-------|\n")
            hws_data = safe_get(sol_data, 'HWS', default={})
            for key, value in hws_data.items():
                md_file.write(f"| {key} | {value} |\n")
            md_file.write("\n")
            
            # Pressure Statistics
            md_file.write("### Pressure (PRE) Statistics\n\n")
            md_file.write("| Statistic | Value |\n")
            md_file.write("|-----------|-------|\n")
            pre_data = safe_get(sol_data, 'PRE', default={})
            for key, value in pre_data.items():
                md_file.write(f"| {key} | {value} |\n")
            md_file.write("\n")
            
            # Most Common Wind Direction
            md_file.write("### Most Common Wind Direction\n\n")
            md_file.write("| Characteristic | Value |\n")
            md_file.write("|---------------|-------|\n")
            wd_most_common = safe_get(sol_data, 'WD', 'most_common', default={})
            for key, value in wd_most_common.items():
                md_file.write(f"| {key} | {value} |\n")
            md_file.write("\n---\n\n")
    
    print(f"Markdown file exported to {output_file}")
    return output_file

# Call the function to export tables
export_sol_tables_to_markdown(response.json())
```
